#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
Notas=[8.5,9.2,7.8,8.9,10]
suma=0
for nota in Notas:
    suma+=nota
print(suma/len(Notas))
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
notamin=Notas[0]
notamax=Notas[0]
for notas in Notas:
    if notamin>notas:
        notamin=notas
    if notamax<notas:
        notamax=notas
print("La nota más baja es "+ str(notamin) + " y la más alta es "+ str(notamax))
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
# Trazar el histograma de precios
plt.figure(figsize=(8, 5))
plt.hist(ds["price"],bins=34)  #ya que la casa más cara cuesta 8.5M he dividido en 34 para que se divida de 250k en 250k
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.title("Histograma de precios de viviendas")
plt.show()
#intente dividir el eje x de 500k en 500k pero no sabía cómo hacerlo solo usando matplotlib.pyplot
```

Hay un claro sesgo en el precio de las casas de los datos proporcionados. Mas de 2/3 de las viviendas están en el rango de precios menores de 500k euros, siendo casi la mitad menor que 250k . En los rangos siguientes (500k-750k-1M) la cantidad de casas baja drasticamente siendo aún más notorios en cúanto subamos más el precio.
